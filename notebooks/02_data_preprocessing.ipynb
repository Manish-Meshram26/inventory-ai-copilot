```python
import pandas as pd 
import numpy as np

#load datasets
calender = pd.read_csv("../data/raw/calendar.csv")
sales_train_val = pd.read_csv('../data/raw/sales_train_validation.csv')
sell_prices = pd.read_csv('../data/raw/sell_prices.csv')
```

```python
print('calender shape:', calender.shape)
print('sales_train_val shape:', sales_train_val.shape)
print('sell_prices shape:', sell_prices.shape)
```

```python
# converting wide sales data to long format so that we can merge it with the other datasets

df = pd.melt(sales_train_val, id_vars=['id', 'item_id', 'dept_id', 'cat_id', 'store_id', 'state_id'], var_name='day', value_name='sales')
```
