# Objective

Create calendar-based features for demand forecasting.

```python
import pandas as pd 

# load datasets
sales_price_df = pd.read_parquet('..\\data\\processed\\03_sales_calendar_prices.parquet')
```

```python
sales_price_df.info()
```

```python
sales_price_df['is_weekend'] = sales_price_df['weekday'].isin(['Saturday', 'Sunday']).astype(int)
```

```python
sales_price_df['is_month_start'] = sales_price_df['date'].dt.is_month_start.astype(int)
sales_price_df['is_month_end'] = sales_price_df['date'].dt.is_month_end.astype(int)
```

```python
sales_price_df[['date', 'is_month_start','is_month_end']].nunique()
```

These features are calendar-based indicators to help the model learn timing effects in demand:

- `is_weekend`: sales often differ on weekends vs weekdays. For many products, customer behavior changes on Saturday/Sunday, so this flag helps the model capture that weekly pattern.
- `is_month_start` and `is_month_end`: demand can shift at the beginning or end of a month because of pay cycles, inventory restocking, promotions, or shopper habits.

Benefits:
- improve forecast accuracy by giving the model explicit temporal signals
- allow it to separate regular calendar effects from other drivers like price or promotions
- useful for models that cannot automatically infer these patterns from raw dates alone
